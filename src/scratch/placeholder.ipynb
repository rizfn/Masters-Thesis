```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot a sine wave with a legend
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x, y, label='sin(x)')
plt.legend()
```

```python
def gaussian(x, mu, sigma):
    return np.exp(-(x - mu)**2 / (2 * sigma**2))

def fast_fourier_transform(x):
    """Compute the discrete Fourier Transform of the 1D array x"""
    x = np.asarray(x, dtype=float)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    M = np.exp(-2j * np.pi * k * n / N)
    return np.dot(M, x)

def inverse_fast_fourier_transform(x):
    """Compute the inverse discrete Fourier Transform of the 1D array x"""
    x = np.asarray(x, dtype=float)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    M = np.exp(2j * np.pi * k * n / N)
    return np.dot(M, x) / N

# plot a gaussian and its fourier transform on subplots
x = np.linspace(-5, 5, 100)
y = gaussian(x, 0, 1)
Y = fast_fourier_transform(y)
plt.subplot(2, 1, 1)
plt.plot(x, y, label='gaussian')
plt.legend()
plt.subplot(2, 1, 2)
plt.plot(x, np.real(Y), label='real')
plt.plot(x, np.imag(Y), label='imag')
plt.legend()
# plot its inverse fourier transform on the same subplot
plt.subplot(2, 1, 1)
plt.plot(x, inverse_fast_fourier_transform(Y), label='inverse')
plt.legend();
```
